```python
import numpy as np
import matplotlib.pyplot as plt

from romshake.simulators.gmpe import GMPE_Simulator
from romshake.core.numerical_rom_builder import NumericalRomBuilder
```

```python
!rm -r test
```

```python
sim = GMPE_Simulator()
bounds = {
    'depth': (0, 60)
}
rb = NumericalRomBuilder(
    'test', sim, 10, 100, 1000, 'halton', bounds, 
    rbf_kernels=['thin_plate_spline'])
```

```python
pred = rb.rom.predict(np.array([[60]]))
plt.imshow(pred['thin_plate_spline'].reshape(30, 30))
plt.colorbar()
```

```python
_, truth = sim.evaluate(rb, np.array([[60]]), write=False)
plt.imshow(truth.reshape(30, 30))
plt.colorbar()
plt.show()
```
